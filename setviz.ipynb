```python
import pandas as pd
from upsetplot import UpSet
import matplotlib.pyplot as plt

# Define data for multiple sets
data = {
    'Set 1': [1, 1, 0, 0, 1, 0],
    'Set 2': [1, 0, 1, 0, 0, 1],
    'Set 3': [0, 1, 0, 1, 0, 0],
    'Set 4': [0, 0, 1, 1, 1, 1]
}

# Create a DataFrame
df = pd.DataFrame(data, index=[f'Item {i+1}' for i in range(len(data['Set 1']))])

# Convert the DataFrame to a multi-index series
set_counts = df.groupby(['Set 1', 'Set 2', 'Set 3', 'Set 4']).size()

# Create UpSet plot
upset = UpSet(set_counts, subset_size='count')
upset.plot()
plt.title("UpSet Plot for Multiple Sets")
plt.show()
```
