```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the objective function
def objective_function(x, y, z):
    return (x - 1)**2 + (y - 2)**2 + (z - 1)**2

# Sphere constraint: x^2 + y^2 + z^2 = r^2
r = 3

# Create a grid of points
theta = np.linspace(0, np.pi, 100)  # azimuthal angle
phi = np.linspace(0, 2 * np.pi, 100)  # polar angle
theta, phi = np.meshgrid(theta, phi)

# Parametric equations for the sphere
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the sphere
ax.plot_surface(x, y, z, color='cyan', alpha=0.5, rstride=5, cstride=5)

# Objective function minimization attempt: we'll use gradient descent and Lagrange multipliers
# For now, let's plot the gradient descent path manually
# Starting point: outside the sphere, near (0, 0, 0)
path_x = np.array([0, 0.5, 1, 1.5, 2, 1.8, 1.7])
path_y = np.array([0, 1, 1.5, 1.8, 2, 2.1, 2])
path_z = np.array([0, 0.5, 1, 1.2, 1.3, 1.1, 1.05])

# Plot the path of gradient descent
ax.plot(path_x, path_y, path_z, marker='o', color='red', label='Gradient Descent Path')

# Add labels to the plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Plot the objective function's minimum on the sphere
ax.scatter(1.7, 2, 1.05, color='green', s=100, label='Minimum (constrained)')

# Add title and legend
ax.set_title('Sphere Constrained Optimization')
ax.legend()

# Show the plot
plt.show()
```
